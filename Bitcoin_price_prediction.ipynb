```python
bit_df_days = bit_df.resample('d').agg({'open': 'first', 'hi': 'max', 
    'lo': 'min', 'close': 'last', 'vol_btc': 'sum',
    'vol_cur': 'sum', 'wp': 'mean', 'ts': 'min'}).iloc[-1000:]
```

```python
#Feature Extration 
X=bit_df_days[['open','hi','lo','vol_btc','vol_cur','wp','ts']]

#Label
y=bit_df_days['close']
```

```python
#Spliting Data into Test and train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.20,random_state=42)
```

```python
#Applying Linear Regression Model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
#Predicting 
y_pred=model.predict(X_test)
print(y_pred)
```

```python
print(f'Cofficients:{model.coef_}')
print(f'intrecept:{model.intercept_}')
```

```python
#Mean_Squared_Error AND  Mean_Absolute_Error
from sklearn.metrics import mean_squared_error, r2_score ,mean_absolute_error
print(f'MSE is {mean_squared_error(y_test, y_pred )}')
print(f'MAE is {mean_absolute_error(y_test, y_pred )}')
```

```python
#Accuracy
accuracy=model.score(X_test,y_test)
accuracy=accuracy*100
accuracy = float("{0:.4f}".format(accuracy))
print('Accuracy is:',accuracy,'%')`
```

```python
#R^2 Score
print(f'r2_score is {r2_score(y_test, y_pred)}')  
```
